```python
filipino_df = myDF[myDF["Race"] == 7] #to check how many people are of Filpino 
```

Part c

```python
valid_age_df = filipino_df[filipino_df["Age"] != 999] #whose age is not equal to 999
```

```python
age_counts = valid_age_df["Age"].value_counts().sort_index() #to count the the each age
```

```python
import matplotlib.pyplot as plt #to make plot we have to import matplotlib
```

#since I wanted to try new approach to make plot, I used GeeksforGeeks website to have more ideas. 

```python
plt.figure(figsize=(12, 6))
plt.bar(age_counts.index, age_counts.values, color='blue', alpha=0.7)
plt.xlabel("Age at Time of Death")
plt.ylabel("Count")
plt.title("Age Distribution at Time of Death for Filipino Individuals")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
myDF = pd.read_csv("/anvil/projects/tdm/data/election/itcont1980.txt", header=None, sep='|')
```

```python
myDF.columns = ["CMTE_ID", "AMNDT_IND", "RPT_TP", "TRANSACTION_PGI", "IMAGE_NUM", "TRANSACTION_TP", "ENTITY_TP", "NAME", "CITY", "STATE", "ZIP_CODE", "EMPLOYER", "OCCUPATION", "TRANSACTION_DT", "TRANSACTION_AMT", "OTHER_ID", "TRAN_ID", "FILE_NUM", "MEMO_CD", "MEMO_TEXT", "SUB_ID"]
#giving columns the heading
```

```python
myDF.head()
```

```python
missing_dt_subset = myDF[myDF['TRANSACTION_DT'].isna()] #to print rows which has missing transaction dt
```

```python
missing_dt_counts = missing_dt_subset['NAME'].value_counts() #to count the name 
```

```python
print("Number of missing TRANSACTION_DT entries per person:")
print(missing_dt_counts) #to print the number of missing people with count
```

```python
#tried a different way to get the similar output.To print the number of missing people with count
```

```python
missing_dt_names = myDF.loc[myDF['TRANSACTION_DT'].isna(), 'NAME'].unique()
```

```python
print("Names with missing TRANSACTION_DT:")
for name in missing_dt_names:
    print(name)
```
