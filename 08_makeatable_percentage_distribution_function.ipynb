```python
def teenagecount (myseries: pd.Series) -> int:
    """
    The teenagecount function takes a Pandas Series as input,
    with any type of data in the Series (for instance, it might be a column of a data frame)
    and returns the number of data points that are between 13 and 19 (inclusive).

    Args:
    myseries (pd.Series): This is any type of Series that we want to study,

    Returns:
    teenagecount (int): This is an integer that shows the number of values between 13 and 19 in the data.
    """
    teenagecount = len(myseries[(myseries >= 13) & (myseries <= 19)])
    return teenagecount
```

```python
def longesttest (mydata: pd.Series) -> str:
    """
    The longesttest function takes a Pandas Series as input,
    containing strings, and returns the longest string.

    Args:
    mydata (pd.Series): This is any type of Series that contains strings which we want to study,

    Returns:
    mylongeststring (str): This is the longest of the strings in the Pandas Series
    """
    mylongeststring = max(mydata, key=len)
    return mylongeststring
```

```python
# Define the function
def mostpopulardate(myseries: pd.Series):
    """
    The mostpopulardate function takes a Pandas Series as input,
    containing dates, and returns the most frequent date along with its occurrence count.

    Args:
    myseries (pd.Series): A Pandas Series containing date values.

    Returns:
    tuple: The most frequent date and its count.
    """
    most_common_date = myseries.value_counts().idxmax()  # Finds the most frequent date
    count = myseries.value_counts().max()  # Finds how many times it occurs
    return most_common_date, count
```

```python
import pandas as pd

# Define the function myaveragedelay
def myaveragedelay(airport_code: str) -> float:
    """
    The myaveragedelay function takes a 3-letter airport code and calculates 
    the average departure delay from the DepDelay column, after removing NA values.

    Args:
    airport_code (str): The 3-letter airport code (e.g., "IND", "JFK").

    Returns:
    float: The average departure delay for flights from the given airport.
    """
    # Load the 1990 flight data
    myDF = pd.read_csv('/anvil/projects/tdm/data/flights/subset/1990.csv')

    # Filter flights by the given airport code
    airport_flights = myDF[myDF["Origin"] == airport_code]

    # Calculate the average departure delay after removing NA values
    avg_delay = airport_flights["DepDelay"].dropna().mean()

    return avg_delay
```

Part b

```python
#to print the average departure delay for IND. Took help from GeeksForGeeks
ind_delay = myaveragedelay("IND")
print(f"Average departure delay for IND: {ind_delay}")
```

```python
#to print the average departure delay for JFK
jfk_delay = myaveragedelay("JFK")
print(f"Average departure delay for JFK: {jfk_delay}")
```
