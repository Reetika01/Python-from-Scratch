```python
import pandas as pd 
#imporitng the library
```

```python
#reading both data frames
productsDF = pd.read_csv("/anvil/projects/tdm/data/icecream/combined/products.csv")
```

```python
reviewsDF = pd.read_csv("/anvil/projects/tdm/data/icecream/combined/reviews.csv")
```

```python
#Merging the data frame on brand &  key
newmergedDF = productsDF.merge(reviewsDF, on=['brand','key'])
```

```python
#Filtering the specific poduct
ChocolateChipCookieDoughDF = newmergedDF[(newmergedDF['brand'] == "bj") & (newmergedDF['key'] == "16_bj")]
```

```python
#to print the result
ChocolateChipCookieDoughDF['stars'].mean()
```

```python
#tried another way with new dataframe
average_stars = ChocolateChipCookieDoughDF['stars'].mean()
```

```python
#to print the result
print(f"The average number of stars for Chocolate Chip Cookie Dough is: {average_stars}")
```

```python
#reading the two dataframes
airportsDF = pd.read_csv("/anvil/projects/tdm/data/flights/subset/airports.csv")
flightdataDF = pd.read_csv("/anvil/projects/tdm/data/flights/subset/1990.csv")
```

```python
#Merging the two dataframes
mybigDF = airportsDF.merge(flightdataDF, left_on="iata", right_on="Origin")
```

```python
# to filter the flights departing from Indiana
indiana_flights = mybigDF[mybigDF['state'] == "IN"]
```

```python
#to print the average departure delay for all flights that have Origin airport in Indiana.
indiana_flights['DepDelay'].mean()
```
