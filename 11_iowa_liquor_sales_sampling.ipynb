```python
import pandas as pd
cols = ['Zip Code', 'Category Name', 'Sale (Dollars)']
```

```python
# Read only the needed columns with correct data types
```

```python
df = pd.read_csv('/anvil/projects/tdm/data/iowa_liquor_sales/iowa_liquor_sales.csv',usecols=cols,dtype={'Zip Code': 'string', 'Category Name': 'string', 'Sale (Dollars)': 'float'})
```

```python
zip_sales = df.groupby('Zip Code')['Sale (Dollars)'].sum().sort_values(ascending=False).head(10)
print(zip_sales)
```

```python
category_sales = df.groupby('Category Name')['Sale (Dollars)'].sum().sort_values(ascending=False).head(10)
print(category_sales)
```

```python
#importing important libraries
import pandas as pd
import numpy as np
```

```python
#Reading only 10th & 15th column from the file, on 0 based indexing
myDF = pd.read_csv("/anvil/projects/tdm/data/election/itcont2020.txt", header=None,sep='|',usecols=[9, 14], encoding='Windows-1252')
```

```python
#renaming the columns
myDF.columns = ["STATE", "TRANSACTION_AMT"]
```

```python
#converting transaction amount to numeric(integer) & missing values or invalid with 0 
myDF['TRANSACTION_AMT'] = pd.to_numeric(myDF['TRANSACTION_AMT'], errors='coerce').fillna(0).astype(np.int64)
```

```python
#Groupby state & sum with the donation amounts
state_sums = (myDF.groupby('STATE')['TRANSACTION_AMT'].sum().sort_values(ascending=False).head(10))
```

```python
#printing the result with the heading
print("Top 10 States by Total Donation Amount (2020):")
print(state_sums)
```

```python
#reading the first 10 rows using tab seperate format
myDF = pd.read_csv('/anvil/projects/tdm/data/movies_and_tv/imdb2024/basics.tsv',sep='\t',nrows=10)
```

```python
#printing the genres column
print(myDF['genres'])
```

```python
#printing the genres with split 
print(myDF['genres'].str.split(','))
```

```python
#exploding it into individual genres
print(myDF['genres'].str.split(',').explode())
```

```python
#counting how many times each genre occured
print(myDF['genres'].str.split(',').explode().value_counts())
```

```python
# Read only the 'genres' column from the full file
full_df = pd.read_csv('/anvil/projects/tdm/data/movies_and_tv/imdb2024/basics.tsv',sep='\t',usecols=['genres'])
```

```python
# Explode and counting all genres from the whole dataset
genre_counts = full_df['genres'].str.split(',').explode().value_counts()
```

```python
# printing the result with the heading
print("Genre Counts Across Entire Dataset:")
print(genre_counts)
```
